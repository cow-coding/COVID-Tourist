```python
# 박물관 전통시장 전시관 생활편의 카페 관광명소 숙박 음식점 촬영장소 유적지
museum = pd.read_csv(PATH+'국내 문화체육관광 분야 박물관 시설 및 소장자료 통계 데이터(방문객 등 통계 포함).csv')
market = pd.read_csv(PATH+'국내 문화체육관광 분야 산업활용용 전통시장 시설 데이터(시군구 행정동 100미터격자코드 포함).csv')
exhibition = pd.read_csv(PATH+'국내 문화체육관광 분야 전시관 시설 데이터(시군구 행정동 100미터격자코드 포함).csv')
life = pd.read_csv(PATH+'국내 문화체육관광시설 인근 생활편의시설 데이터.csv')
cafe = pd.read_csv(PATH+'국내 문화체육관광시설 인근 카페 및 디저트 판매시설 데이터.csv')
hotplace = pd.read_csv(PATH+'국내 지역별 관광명소데이터(시군구 행정동 100미터격자코드 포함).csv')
house = pd.read_csv(PATH+'국내 지역별 문화체육관광시설 인근 숙박시설 데이터.csv')
restaurant = pd.read_csv(PATH+'국내 지역별 문화체육관광시설 인근 음식점(한식, 일식, 양식, 외국식 등) 데이터.csv')
movie = pd.read_csv(PATH+'국내 지역별 영화 드라마 촬영장소 및 시설 데이터(시군구 행정동 100미터격자코드 포함).csv')
oldplace = pd.read_csv(PATH+'국내 지역별 유적지(고궁,성,문,고택,생가,민속마을,유적사적 등) 데이터.csv')
```

```python
jj_musesum = museum[(museum['ctprvn_nm'] == '전라북도') & (museum['sgnr_nm'].str.contains('전주'))]
jj_market = market[(market['CTPRVN_NM'] == '전라북도') & (market['SGNR_NM'].str.contains('전주'))]
jj_exhibion = exhibition[(exhibition['sido_nm'] == '전라북도') & (exhibition['sgg_nm'].str.contains('전주'))]
jj_life = life[(life['sido_nm'] == '전라북도') & (life['sgg_nm'].str.contains('전주'))]
jj_cafe = cafe[(cafe['sido_nm'] == '전라북도') & (cafe['sgg_nm'].str.contains('전주'))]
jj_hotplace = hotplace[(hotplace['sido_nm'] == '전라북도') & (hotplace['sgg_nm'].str.contains('전주'))]
jj_house = house[(house['sido_nm'] == '전라북도') & (house['sgg_nm'].str.contains('전주'))]
jj_restaurant = restaurant[(restaurant['sido_nm'] == '전라북도') & (restaurant['sgg_nm'].str.contains('전주'))]
jj_movie = movie[(movie['sido_nm'] == '전라북도') & (movie['sgg_nm'].str.contains('전주'))]
jj_oldplace = oldplace[(oldplace['sido_nm'] == '전라북도') & (oldplace['sgg_nm'].str.contains('전주'))]
```
