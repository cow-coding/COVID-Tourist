## 전주 관광객 데이터 분석

```python
tourist_si_day['I_CNT'] = tourist_si_day['I_CNT'].apply(lambda x: int(x.replace(',','')) if type(x) == str else int(x))
tourist_si_day['O_CNT'] = tourist_si_day['O_CNT'].apply(lambda x: int(x.replace(',','')) if type(x) == str else int(x))
tourist_si_day['F_CNT'] = tourist_si_day['F_CNT'].apply(lambda x: int(x.replace(',','')) if type(x) == str else int(x))
tourist_si_day = tourist_si_day.set_index(tourist_si_day['ETL_YMD'])
tourist_day_2019 = tourist_si_day[('2019-03' <= tourist_si_day['ETL_YMD']) & (tourist_si_day['ETL_YMD'] < '2020-01')]
tourist_day_2020 = tourist_si_day[('2020-01' <= tourist_si_day['ETL_YMD']) & (tourist_si_day['ETL_YMD'] < '2021-01')]
tourist_day_2021 = tourist_si_day[('2021-01' <= tourist_si_day['ETL_YMD'])]
```

### 2019

```python
tourist_jj_2019_week = pd.DataFrame(columns=['기준일자', '현지인_관광객', '외지인_관광객', '외국인_관광객' ,'시군구코드', '시도코드'])

tourist_jj_2019_week['현지인_관광객'] = tourist_day_2019['I_CNT'].resample('1W').sum()
tourist_jj_2019_week['외지인_관광객'] = tourist_day_2019['O_CNT'].resample('1W').sum()
tourist_jj_2019_week['외국인_관광객'] = tourist_day_2019['F_CNT'].resample('1W').sum()

tourist_jj_2019_week['기준일자'] = tourist_jj_2019_week.index

tourist_jj_2019_week['시군구코드'] = tourist_si_day['SGG_CD'][1]
tourist_jj_2019_week['시도코드'] = tourist_si_day['SIDO_CD'][1]

tourist_jj_2019_week['총_관광객'] = tourist_jj_2019_week[['외지인_관광객', '외국인_관광객']].apply(lambda x : x['외지인_관광객'] + x['외국인_관광객'], axis=1)
```

```python
tourist_jj_2019_month = pd.DataFrame(columns=['기준일자', '현지인_관광객', '외지인_관광객', '외국인_관광객' ,'시군구코드', '시도코드'])

tourist_jj_2019_month['현지인_관광객'] = tourist_day_2019['I_CNT'].resample('1M').sum()
tourist_jj_2019_month['외지인_관광객'] = tourist_day_2019['O_CNT'].resample('1M').sum()
tourist_jj_2019_month['외국인_관광객'] = tourist_day_2019['F_CNT'].resample('1M').sum()

tourist_jj_2019_month['기준일자'] = tourist_jj_2019_month.index

tourist_jj_2019_month['시군구코드'] = tourist_si_day['SGG_CD'][1]
tourist_jj_2019_month['시도코드'] = tourist_si_day['SIDO_CD'][1]

tourist_jj_2019_month['총_관광객'] = tourist_jj_2019_month[['외지인_관광객', '외국인_관광객']].apply(lambda x : x['외지인_관광객'] + x['외국인_관광객'], axis=1)
```
